```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn(x) :
    y = 4*x**2 - 2*x + 3
    return y
def derF(x):
    return 8*x-2
```

```python
x = np.linspace(-3.5,4,100)
y = fn(x)
plt.plot(x,y)
plt.show()
```

```python
x = -3
alpha = 0.1
xs = [x]
```

```python
for i in range(10):
    x = x- alpha*derF(x)
    xs.append(x)
```

```python
x = np.linspace(-3.5,4,100)
y = fn(x)
plt.plot(x,y)
plt.plot(xs,fn(np.array(xs)),'r')
plt.xlim(-4,5)
plt.ylim(0,50)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2*x**2 + 3*x + 5
```

```python
X = np.linspace(-4,3,1000)
y = f(X)
```

```python
plt.plot(X,y)
plt.show
```

```python
def det_fx(x):
    return 4*x + 3
```

```python
alpha = 0.50
x = 2
epochs = 10
xs = [x]
for e in range(epochs):
    x = x - alpha*det_fx(x)
    xs.append(x)
x
```

```python
plt.plot(X,y)
plt.plot(xs, f(np.array(xs)), c = 'r') 
plt.show()
```

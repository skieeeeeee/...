```python
data = pd.read_csv("Titanic.csv")
```

```python
data["Mod_Age"] = data["Age"].fillna(data['Age'].mean())
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns = ['Cabin'])
data['Embarked'] = data['Embarked'].dropna()
data = data.drop(index = data[data['Embarked'].isnull()].index)
```

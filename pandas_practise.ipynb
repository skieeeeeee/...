```python
data = {
    "name" : [ "spongebob" , "patrick" , "squidward"],
    "age" : [ 30,25,35 ]
    }
```

```python
df = pd.DataFrame(data , index = ["employe 1" , "employe 2" , "employe 3"])
df
```

```python
df.loc["employe 1"]
```

```python
# add a new column 
df["job"] = ["cook" , "N/A" , "cashier"]
```

```python
df
```

```python
# add a new row 
new_rows = pd.DataFrame([{"name" : "sandy" , "age" : 45 , "job" : "engineer" }, 
                        { "name" : "mandy" , "age" : 10 , "job" : "doctor" }],
                        index = ["employe 4" , "employe 5 "])
```

```python
df = pd.concat([df , new_rows])
```
